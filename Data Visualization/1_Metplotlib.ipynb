```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Matplotlib Basics
```

```python
x = np.linspace(0, 5, 11)
x

y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("Graph")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
```

```python
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.subplot(2, 2, 2)
plt.plot(x**2, x**2)
plt.subplot(2, 2, 3)
plt.plot(y,x)
plt.subplot(2, 2, 4)
plt.plot(x,y)
```

```python
# Object Oriented way

# Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

# figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).
```

```python
fig = plt.figure(figsize=(20,8), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])
axis2.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y)
plt.savefig("BasicPlot.png")
```
