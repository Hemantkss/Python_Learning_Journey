```python
import numpy as np 
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr, 25)
print("Q1 is : ",Q1)
Q2 = np.percentile(arr, 50)
print("Q1 is : ",Q2)
Q3 = np.percentile(arr, 75)
print("Q3 is : ",Q3)
```

```python
IQR = Q3 - Q1
print("IQR is : ",IQR)
```

```python
UpperFence = Q3 + (1.5 * IQR)
print("UpperFence is : ",UpperFence)
LowerFence = Q1 - (1.5 * IQR)
print("LowerFence is : ",LowerFence)
```

```python
l = []

for i in arr:
    if i >= LowerFence and i <= UpperFence:
        l.append(i)

arr2 = np.array(l)
        
        
```

```python
print(arr)
print(arr2)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x = arr2)
```

```python
# Add vertical lines and labels
for val, label in zip([LowerFence, Q1, Q2, Q3, UpperFence],
                      ["Lower Fence", "Q1", "Median (Q2)", "Q3", "Upper Fence"]):
    plt.axvline(val, color='red', linestyle='--')
    plt.text(val, 0.05, f'{label}\n{val:.2f}', rotation=90,
             ha='center', va='bottom', color='black')

plt.title("Boxplot with IQR, Median and Fences")
plt.yticks([])  # Hide y-axis
plt.tight_layout()
plt.show()
```
